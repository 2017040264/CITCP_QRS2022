# 内容提取

```python
def analysis_noth(df):
    
    reward_list=["tcfail","add_tcfail_1","add_tcfail_2","add_tcfail_3"]
    exlist=["Dataset"]+reward_list*4
    lista=[] 
    for column, env in enumerate(sorted(df['env'].unique())):
        env1=env.title()
        env1=env1.replace("_"," ")
        if env1=="Iofrol":
            env1="IOF/ROL"
        slist=[env1]
            #for re in ["APHF","failcount","tcfail","timerank"]:
        for title in ['napfd','recall','ttf','durations']:
            for re in reward_list:
            
                tdf=df[df["rewardfun"].isin([re]) & df["env"].isin([env])]
                x=str(tdf[title]).split()[1]
                #print(x)
                slist.append(float(x))
                
        lista.append(slist)
    
    

    
    #print(exlist)
    # list转dataframe
    cdf = pd.DataFrame(lista, columns=exlist)

    # 保存到本地excel
    cdf.to_excel("qr2_.xlsx", index=False)

        
    
```

```python
analysis_noth(df)
```
